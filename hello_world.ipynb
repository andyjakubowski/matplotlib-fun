## Object-oriented, imperative style vs. more declarative `pyplot` interface

### Object-oriented

The so called object-oriented style lets you explicitly create figures and axes, and call methods on them.

Doing things this way is more powerful, but slightly more complex.

```python
x = np.linspace(0, 2, 100)

fig, ax = plt.subplots()
ax.plot(x, x, label='linear')
ax.plot(x, x**2, label='quadratic')
ax.plot(x, x**3, label='cubic')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title("Simple Hello World plot")
ax.legend()
```

### Declarative

The other option is to tell `pyplot` what you want, and let it automatically create and manage the figures and axes.

This way of doing things is simpler, and MATLAB-like. It’s intended for interactive plots and simple cases of programmatic plot generation.

```python
x = np.linspace(0, 2, 100)

plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Hello World plot")
ax.legend()
```
